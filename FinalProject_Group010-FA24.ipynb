## Background and Prior Work

Urbanization, or the percent of urban population, refers to how much of a country's population lives in an urban (city) area. We are using this as a measure of how urbanized a country is. A country's happiness level index is a 1 to 10 scale which measures how "happy" a country is, taking into account factors like life satisfaction, psychological well-being, health, community, and more.

The World Happiness Report has been used as a measure of determining the world's happiest countries by analzing six particular categories: gross domestic product (GDP) per capita, social support, healthy life expectancy, freedom to make life choices, generosity of the general population, and perceptions of internal and external corruption levels.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) In recent years, researchers have investigated how urbanization, GDP, government types, and life expectancy correlate with happiness levels. Findings are still varied, however, with some studies showing positive correlations and others suggesting a complex lack of correlation. Findings from the report indicate that indeed, in many countries, happiness levels are higher in urban areas, but this trend is nuanced. For example, a comparison in the level of happiness between urban and rural populations in poorer regions, such as Sub-Saharan Africa, indicates higher levels of happiness among urban dwellers due to increased job opportunities and more service and infrastructural provisions, while some Western countries tend to show higher well-being indicators in rural areas due to less stress and closer communal bonding.

The idea of how a country's GDP can affect things such as quality of living has been researched extensively. More specific to our question is the idea of how living in urban, rural, or suburban areas can affect someone's happiness. There has been much research around this already, pointing to the idea that people are more happy living in suburban "sweet spots" than they living in rural or urban areas.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) We are researching an intersection between these: rather than researching a country's GDP or something similar and comparing it to their happiness index score, we are researching how the urbanization of a country can correlate to the happiness index score of that country. Do more urbanized countries tend to be happier? If individuals are happier living in suburban areas, does that mean that moderately urbanized countries will be happiest?

In a study in 2020 conducted by the European Commission and Gallup analyzed the degree of urbanization's effect on happiness. This study asked 360,000 people across cities, towns, semi-dense areas, and rural areas of 117 countries to rate their lives on a scale from 0 to 10. The study found that on average, the percent of people in any country who gave their lives a score between 8 and 10 is 2-4% higher in cities than rural areas. The study recognizes that a variety of other factors could have contributed to this difference in happiness level between cities and rural areas. One of these factors they measured was access to financial resources. Differences in access to financial resources are most pronounced in low-income economies as the study found that 18% of rural residents reported having a bank account, while 30% of city residents reported having a bank account. Additionally, the study found that rural residents in lower-income economies were more likely than city residents to lack having enough money to buy food. Finally, the study measured support from friends or relatives to help them in times of need. On average, the rural residents were less likely to report having support in times of need than city dwellers.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Overall, this study depicts how complex the idea of measuring happiness is as it cannot be narrowed down to a singular factor like degree of urbanization.

One key factor in the changing of happiness levels is the amount of greenery sustained during the changing from rural environments to urban environments. As urban population increases, access to green spaces and quality of ecosystems decrease, making this a topic of concern among urban planners. Jorge E. Patino studied one of the most urbanized regions in the world, Latin America and the Caribbean, where his discovered a positive correlation between access to urban green spaces and happiness levels in urban areas.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) There are however several other factors that need to be considered when studying the effects of urban green spaces and happiness levels. Patino also notes that in some cities, those with higher socioeconomic status access to better quality urban green spaces, which suggests that socioeconomic status could be a confounding variable that must be studied further. Additionally, he found that reports of happiness tended to be higher the older the respondant was. We want to discover exactly what factors, such as access to green spaces, are effected by urbanization, and how these factors contribute to happiness levels in urban populations.

Overall, while degree of urbanization may be a significant factor in determining the happiness level of people in a certain country, there are a variety of variables that contribute to determining the level of happiness for a population. Prior research suggests that happiness is a complex concept to measure, and variables such as GDP, government type, average life expectancy etc. can each play a role in the complexity of determining the happiness level for a country. Despite the amount of research done on this topic, determining if there is the correlation between a country's percent of urban population and their happiness level still remains to be answered. Furthermore, understanding the relationship that the aforementioned variables have with happiness level is an important aspect of this research topic to explore.

## Exploring the Relationship Between Happiness and Urbanization Through Choropleth Maps
The dual choropleth map visualization was chosen to provide a global perspective on the relationship between the Happiness Index (Ladder Score) and Urban Population Percentage. We aim to see the regions with greater Urban Population Percentages have a higher Happiness Index (Ladder Score). The Happiness Index map highlights regional patterns of well-being, while the Urban Population map illustrates the distribution of urbanization across countries. Juxtaposing these maps allows for visual comparisons to identify correlations or discrepancies between happiness and urbanization on a global scale. Choropleth maps are particularly effective for displaying geographical data, using color gradients to intuitively represent data intensity. This approach offers a clear and accessible way to explore spatial patterns, identify outliers, and set the stage for deeper analysis of the interplay between happiness, urbanization, and other influencing factors.

```python
## This code block visualizes global data with two chloropleth maps:
## 1. Happiness Index by Country (Ladder score)
## 2. Urban Population % by Country
## Steps: Load shapefile, match country names, merge data, and plot data

# -- This code block is completely generated by ChatGPT with the prompt:
# "use geopandas package to write code that makes two choropleth maps to display
# the Happiness Index by Country and the Urban Population % by Country using the 
# data in the merged_df dataset" --
# Load shapefile data
shapefile_path = "datasets/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp"
world = gpd.read_file(shapefile_path)

# Get list of countries in the shapefile
shapefile_countries = world['ADMIN'].unique()

# Function to map country names
# This is used to map country names from the merged DataFrame to the shapefile
# -- The code of this function is completely generated by ChatGPT with the prompt:
# "write a python function that maps country names from the merged_df dataset to 
# the country names in the shapefile data. For example "United States" in the 
# merged_df should be mapped to "United States of America" in the shapefile" --
def match_country(name):
    match = process.extractOne(name, shapefile_countries, scorer=fuzz.partial_ratio)
    if match and match[1] > 80:
        return match[0]
    return None

# Map DataFrame country names to shapefile names
merged_df['Mapped Country'] = merged_df['Country name'].apply(match_country)

# Merge data
world = world.merge(merged_df, how="left", left_on="ADMIN", right_on="Mapped Country")

# Set style
sns.set_theme(style="whitegrid")
# Background color for the plot area
plt.rcParams['font.family'] = 'Georgia'
plt.rcParams['axes.facecolor'] = '#f5f5f5'  

# Create subplots for both maps
fig, ax = plt.subplots(1, 2, figsize=(20, 12))

# Happiness Index Choropleth 
world.boundary.plot(ax=ax[0], linewidth=0.8, color='black')
world.plot(
    column="Ladder score",  # Change this to your happiness index column
    ax=ax[0],
    legend=True,
    legend_kwds={'label': "Happiness Index", 'orientation': "horizontal"},
    cmap='YlOrRd',
    missing_kwds={"color": "lightgrey", "label": "No data"}
)
ax[0].set_title("Happiness Index by Country", fontsize=18, fontweight='bold', color='black')
ax[0].axis('off')

# Urban Population % Choropleth 
world.boundary.plot(ax=ax[1], linewidth=0.8, color='black')
world.plot(
    column="Urban population (% of total population)",  # Change to your column
    ax=ax[1],
    legend=True,
    legend_kwds={'label': "Urban Population %", 'orientation': "horizontal"},
    cmap='Blues',
    missing_kwds={"color": "lightgrey", "label": "No data"}
)
ax[1].set_title("Urban Population % by Country", fontsize=18, fontweight='bold', color='black')
ax[1].axis('off')

# Customize layout and show plot
plt.tight_layout()
plt.show()
```

## Looking at Linear Regression Plot for Ladder Score(Happiness Index) vs Urban Population Percent
The linear regression plot was chosen to visualize and assess the relationship between the Happiness Index (Ladder Score) and the percentage of the urban population. This plot provides a clear depiction of how these two variables are correlated, helping us identify whether countries with higher happiness scores tend to have a larger urban population or if the relationship is negligible. The regression line overlays a trend that summarizes the general direction of this relationship, offering an intuitive way to interpret the data. We used scatter points to represent individual data entries, which gives a sense of variability and distribution around the trend. The linear regression line and scatter points work together to highlight both the overall trend and the deviations, offering insights into the strength and consistency of this association. This visualization is a critical step in evaluating whether urbanization plays a significant role in influencing happiness, informing further analyses on causality and other contributing factors.

```python
## This code block visualizes the relationship between Urban Population % and Happiness Index:
## 1. Scatter plot showing data points for each country
## 2. Regression line indicating the overall trend
## Steps: Set plot styling, load data, plot scatter points with regression line, and format the visualization

# Set Seaborn theme and font for consistent styling
sns.set_theme(style='whitegrid')
plt.rcParams['font.family'] = 'Georgia'
plt.rcParams['axes.facecolor'] = '#f5f5f5' 


# Create a scatter plot with regression line
plt.figure(figsize=(10, 6)) # Set the figure size
sns.regplot(
    x="Urban population (% of total population)", # X-axis: Urban population percentage
    y="Ladder score",  # Y-axis: Happiness index (Ladder score)
    data=merged_df, 
    line_kws={"color": sns.color_palette("Blues")[4], "linewidth": 2},  # Color of the regression line
    scatter_kws={"alpha": 0.6, "color": sns.color_palette("Blues", 1)[0]}  # Transparency of the scatter points
)

# Add axis labels and plot title
plt.xlabel("Urban Population (% of total population)", fontsize=14, labelpad=10) # X-axis label
plt.ylabel("Happiness Index (Ladder score)", fontsize=14, labelpad=10) # Y-axis label
plt.title("Happiness Index vs Urban Population %", 
          fontsize=18, fontweight='bold', pad=15) # Plot title

# Add grid lines and show plot
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

The scatter plot illustrates the relationship between the Happiness Index (Ladder Score) and the Urban Population percentage of the total population. ## Comprehensive OLS Regression Analysis with Confounding Factors
Ordinary Least Squares (OLS) regression was applied to the dataset for all countries to quantify the relationship between the urban population percentage and happiness (Ladder Score) at a global scale. This method establishes an understanding of whether urbanization alone is a significant predictor of happiness across diverse national contexts. By isolating the urban population percentage as a predictor, the model evaluates its independent contribution to happiness without interference from other variables. This approach effectively identifies the existence and strength of a linear relationship between urbanization and happiness on a global scale.


The P-value in the OLS regression output quantifies the probability of observing the data—or something more extreme—under the assumption that the null hypothesis is true. In this analysis, the null hypothesis typically states that there is no relationship between the predictor variable(s) (e.g., urban percentage, GDP) and the outcome variable (Ladder Score). If the P-value is less than the chosen significance level (α=0.05), we reject the null hypothesis, concluding that the predictor variable(s) have a statistically significant relationship with the outcome variable.

```python
## This code block performs Ordinary Least Squares (OLS) regression to examine the relationship between 
## urban population percentage and happiness (ladder score)
## Steps: Select relevant columns, rename for clarity, performs OLS regression, and display the results

# Select releveant data for OLS analysis
all_countries_ols = merged_df[['Urban population (% of total population)', 
                               'Ladder score', 
                               'Log GDP per capita', 
                               'Proportion of Urban Population living in Slums', 
                               'Life expectancy', 
                               'Social support']]

# Rename columns for clarity in regression analysis
all_countries_ols = all_countries_ols.rename(columns={'Urban population (% of total population)': 'urban_percent',
                                                      'Ladder score': 'ladder_score',
                                                      'Log GDP per capita': 'log_gdp',
                                                      'Proportion of Urban Population living in Slums': 'slums',
                                                      'Life expectancy': 'life_expectancy',
                                                      'Social support': 'social_support'})

# Perform OLS regression with 'ladder_score' as the dependent variable 
# and 'urban_percent' as the independent variable
outcome_1, predictors_1 = patsy.dmatrices('ladder_score ~ urban_percent', all_countries_ols)

# Fit the OLS model and output OLS regression results
mod_1 = sm.OLS(outcome_1, predictors_1)
res_1 = mod_1.fit()
print(res_1.summary())
```

From this model, we can see that the P value is 0.00, which means that we see a significant correlation between urban population percent and happiness levels, suggesting that urban population percentage does influence a country’s happiness, like we predicted. 


However, we want to look at this relationship further, as the scatter matrix we created seems to show that all independent variables we would like to test have a correlation with a country's ladder score. As such, it is important to address whether or not our statistically significant value for urban percent isn't just a result of confounding variables that we suggested. 


So, OLS regression was extended to include additional predictors—such as GDP per capita, life expectancy, social support, and the proportion of the urban population living in slums. Incorporating these confounders allows the model to disentangle the effects of urbanization from other socio-economic and demographic factors that are likely correlated with both urbanization and happiness. This multi-variable regression provides a more nuanced understanding of the urbanization-happiness relationship by determining whether urban population percentage remains a significant predictor of happiness when these confounders are considered. Additionally, it reveals the relative contributions of each factor to the overall variance in happiness, ensuring a more robust and accurate representation of the dynamics at play.

```python
## This code block performs Ordinary Least Squares (OLS) regression to examine the relationship between 
## happiness (ladder score) and multiple predictors: urban population percent, GDP, urban slums population
## percent, life expectancy, and social support
## Steps: Define the dependent and independent variables, perform OLS regression, and display the result

# Perform OLS regression with 'ladder_score' as the dependent variable
# and 'urban_percent', 'log_gdp', 'slums', 'life_expectancy', 'social_support' as independent variables
outcome_2, predictors_2 = patsy.dmatrices('ladder_score ~ urban_percent + log_gdp + slums + life_expectancy + social_support', all_countries_ols)

# Fit the OLS model and output OLS regression results
mod_2 = sm.OLS(outcome_2, predictors_2)
res_2 = mod_2.fit()
print(res_2.summary())
```

From this model we can see that all of our independent variables have a statistically significant effect on the ladder score except for slums, which has a p-value of 0.494, way over our threshold of 0.05. As such, this proves that slums is a confounding variable in determining ladder score, as though it might see to be correlated with ladder score on a scatter plot, it is due to its correlation with the other independent variables that actually affect ladder score.

Looking at the t-values for each independent variable, we can also glean just how statistically significant each is, thus allowing us to see which factors have the most impact on the ladder score. A higher t-value represents more standard deviations away from a normal mean that could be achieved randomly. As such, social support has the most effect on the ladder score, with a t-value of 11.58, and GDP per capita comes in second with a t-value of 8.127.

Although urban population percent ranks last among our other factors it still is a statistically significant factor in determining ladder score. This means that our initial concerns of GDP being a confound can be put to rest, as urbanization plays its own role in determining a country's happiness levels, despite still being heavily correlated with GDP.
